#lab01 introduction to python library

```python
import numpy as np
list1=[1,2,3]
array1=np.array(list1)
array1
```

```python
list2=[4,5,6]
array2=np.array(list2)
array2
```

```python

print("add:",array1+array2)


print("sub:",array1-array2)
print("mul:",array1*array2)
print("florr:",array1%array2)

print("div:",array1/array2)
print("power:",array1**array2)
#sine function

print("sine1",np.sin(array1))

print("sine2",np.sin(array2))


#logorithm
print("natural ",np.log(array1))
print("base10 ",np.log10(array1))
print("base2 ",np.log2(array1))

#exponential
print("exponential",np.exp(array1))
```

Lab 1 part-02 lab task is to visualize the datasets or data using Matplolib

```python
people=['Vyshu','Satya','Nandu','Reethu','Pooja']
age=[18,18,17,19,18]
weight=[50,49,56,30,47]
height=[160,125,155,187,128]

import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
plt.scatter(weight, height)
plt.title("Relationship between weight and height of patients")
plt.ylabel("height in metres")
plt.xlabel("weight of patients in kgs")
plt.show()
```

Generate Bar Plot

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=16,fontstyle='italic')

#main plot function 'bar'
plt.bar(x=people,height=weight,width=0.6,color='purple',edgecolor='b',alpha=0.6)
plt.xlabel("people",fontsize=15)

#plt.xticks(fontsize=14,rotation=30)
#plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))

#main plot function 'hist'
plt.hist(weight,color='red',edgecolor='b',alpha=0.75,bins=5)

plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("weight in kgs",fontsize=15)
#plt.xticks(fontsize=15)
#plt.yticks(fontsize=15)
plt.show()
```
